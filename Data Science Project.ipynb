```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('walmart_sales.csv')

df.head()
```

#Q1. What is the total sales?

```python
print(df['Weekly_Sales'].sum())
```

#Q2. What is the average weekly sales?

```python
print(df['Weekly_Sales'].mean())
```

#Q3. What is the highest weekly sales?

```python
print(df['Weekly_Sales'].max())
```

#Q4. What is the lowest weekly sales?

```python
print(df['Weekly_Sales'].min())
```

#Q5. How many records are present in the dataset?

```python
print(df.shape[0])
```

#Q6. How many columns are present?

```python
print(df.shape[1])
```

#Q7. How many unique stores are present?

```python
print(df['Store'].nunique())
```

#Q8. Which store has the highest total sales?

```python
print(df.groupby('Store')['Weekly_Sales'].sum().idxmax())
```

#Q9. Find mean sales using NumPy.

```python
print(np.mean(df['Weekly_Sales']))
```

#Q10. Find median sales using NumPy.

```python
print(np.median(df['Weekly_Sales']))
```

#Q11. Find standard deviation of sales.

```python
print(np.std(df['Weekly_Sales']))
```

#Q12. Find 90th percentile of weekly sales.

```python
print(np.percentile(df['Weekly_Sales'],90))
```

#Q13. Create NumPy array of Weekly Sales.

```python
sales_array=np.array(df['Weekly_Sales'])
print(sales_array)
```

#Q14. Show summary statistics.

```python
print(df.describe())
```

#Q15. Plot histogram of Weekly Sales.

```python
plt.hist(df['Weekly_Sales'])
plt.title("Weekly Sales Distribution")
plt.show()
```

#Q16. Plot bar chart of Top 10 Stores by Sales.

```python
df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.show()
```

#Q17. Plot scatter chart of Temperature vs Weekly Sales.

```python
plt.scatter(df['Temperature'],df['Weekly_Sales'])
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.show()
```

#Q18. Create Seaborn Boxplot of Weekly Sales.

```python
sns.boxplot(x=df['Weekly_Sales'])
plt.show()
```

#Q19. Create Seaborn Countplot of Holiday Flag.

```python
sns.countplot(x='Holiday_Flag',data=df)
plt.show()
```

#Q20. Create Correlation Heatmap.

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.show()
```
