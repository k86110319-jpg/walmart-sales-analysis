```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('walmart_sales.csv')

df.head()
```

## Q1. What is the total sales?

```python
total_sales = df['Weekly_Sales'].sum()

print("Total Sales =", total_sales)
```

## Q2. What is the average weekly sales?

```python
average_sales = df['Weekly_Sales'].mean()

print("Average Weekly Sales =", average_sales)
```

#Q3. What is the highest weekly sales?

```python
max_sales = df['Weekly_Sales'].max()

print("Highest Weekly Sales =", max_sales)
```

#Q4. What is the lowest weekly sales?

```python
lowest_sale = df.loc[df['Weekly_Sales'].idxmin()]

print(lowest_sale)
```

#Q5. How many unique stores are present in the dataset?

```python
total_stores = df['Store'].nunique()

print("Total Unique Stores =", total_stores)
```

#Q6. Which store generated the highest total sales?

```python
store_sales = df.groupby('Store')['Weekly_Sales'].sum()

highest_store = store_sales.idxmax()
highest_sales = store_sales.max()

print("Store with Highest Sales:", highest_store)
print("Total Sales:", highest_sales)
```

#Q7. Which store generated the lowest total sales?

```python
store_sales = df.groupby('Store')['Weekly_Sales'].sum()

lowest_store = store_sales.idxmin()
lowest_sales = store_sales.min()

print("Store with Lowest Sales:", lowest_store)
print("Total Sales:", lowest_sales)
```
